```python
import pandas as pd
df=pd.read_csv('/content/heart-disease/heart_disease_uci.csv')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df.isnull().sum()
```

```python
numeric_cols=df.select_dtypes(include='number').columns
df[numeric_cols]=df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df[numeric_cols].hist(figsize=(15,20))
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Numeric Feature Correlation')
plt.show()
```

```python
cat_cols =df.select_dtypes(include='object').columns.tolist()  #one-hot encode all 'objects' categorical columns
if 'num' in cat_cols:
  cat_cols.remove('num') #don't encode target
```

```python
x=df.drop('num', axis=1)
y=(df['num'] > 0).astype(int)
```

```python
x = pd.get_dummies(x, columns=cat_cols)
print("final feature columns:", x.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
y_pred_lr = lr_model.predict(x_test_scaled)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d' , cmap='Blues')
plt.title('confusion Matrix (Logistic Regression)')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train_scaled, y_train)
y_pred_rf = rf_model.predict(x_test_scaled)
```

```python
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
feat_imp = pd.Series(rf_model.feature_importances_, index=x.columns)
feat_imp.nlargest(10).plot(kind='barh')
plt.title('Random forest Feature Importance')
plt.show()
```

```python
import joblib
joblib.dump(rf_model, 'heart_rf_model_pkl')
joblib.dump(scaler, 'heart_scaler.pkl')
```

```python
import joblib
import pandas as pd

user_df = pd.read_csv('heart_dataset.csv')

#getting columns list from training dataaframe
numeric_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()
bool_cols = df.select_dtypes(include='bool').columns.tolist()

#dropping columns which are extra in user_df than required to avoid error
numeric_cols = [col for col in numeric_cols if col in user_df.columns]
cat_cols = [col for col in cat_cols if col in user_df.columns]
bool_cols = [col for col in bool_cols if col in user_df.columns]


#fill the missing numeric column and cat column
user_df[numeric_cols]=user_df[numeric_cols].fillna(user_df[numeric_cols].mean())

for col in cat_cols:
  user_df[col]=user_df[col].fillna('Unkhown')

for col in bool_cols:
  user_df[col]=user_df[col].astype(int)


#one-hot encoding cat columns
user_df_encoded=pd.get_dummies(user_df,columns=cat_cols)

#allign columns
user_df_encoded = user_df_encoded.reindex(columns=x.columns, fill_value=0)

#scale data
scaler = joblib.load('heart_scaler.pkl')
user_df_scaled = scaler.transform(user_df_encoded)

#prediction
prediction = joblib.load('heart_rf_model_pkl')
preds=prediction.predict(user_df_scaled)
user_df['Heart_Disease_Prediction']=preds

#show result
print(user_df)
```
